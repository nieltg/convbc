```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
 
# data to plot
n_groups = 5
means_frank = (np.average([1.09, 1.07, 1.14]), np.average([2.60, 2.61, 2.60]), np.average([2.74, 2.62, 2.59]), np.average([2.62, 2.60, 2.69]), np.average([1.15, 1.14, 1.09]))
means_guido = (np.average([1.11, 1.15, 1.08]), np.average([1.14, 1.13, 1.12]), np.average([1.14, 1.10, 1.14]), np.average([2.68, 2.69, 2.66]), np.average([1.22, 1.18, 1.22]))
 
# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8
 
rects1 = plt.bar(index, means_frank, bar_width,
alpha=opacity,
color='b',
label='Enkripsi')
 
rects2 = plt.bar(index + bar_width, means_guido, bar_width,
alpha=opacity,
color='g',
label='Dekripsi')
 
plt.xlabel('Mode')
plt.ylabel('Waktu Eksekusi (detik)')
plt.title('Perbandingan Waktu Eksekusi Berbagai Mode (4 Kb)')
plt.xticks(index + bar_width/2, ('ECB', 'CBC', 'CFB', 'OFB', 'CTR'))
plt.legend()
 
plt.tight_layout()
plt.show()
```
